```python
df_new = df_new[df_new['language'] == 'en']
X = df_new['text']
```

```python
vectorize = TfidfVectorizer(ngram_range=(1,2),max_features=20000,stop_words=sws)
tfidf = vectorize.fit_transform(X)
```

```python
tfidf_df = pd.DataFrame(tfidf.toarray(),columns = vectorize.get_feature_names(),index=X.index)
y = df_new['ratings_overall']
X = pd.concat([tfidf_df,df_new[['word_count','punctuation_count','word_density','uppercase_ratio']]],
              join='inner',axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,y)
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```
